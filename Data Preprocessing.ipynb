```python
# count of each director names
dataset['director'].value_counts()
```

From the above result, we can see that the names of director whose name were not given are 2588.
Since they are not given, they have to be removed.

```python
# drop rows in which the director name is not given
dataset = dataset.where(dataset['director'] !='Not Given').dropna()
```

```python
dataset
```

```python
# count of each director names
dataset['country'].value_counts()
```

From the above result, we can see that that 287 country names were not given.
Since they are not given, they also need to be removed.

```python
# drop rows in which the country value is not given
dataset = dataset.where(dataset['country']!='Not Given').dropna()
```

```python
dataset
```

```python
# print out the data type of each column
dataset.dtypes
```

```python
dataset['date_added'] = pd.to_datetime(dataset['date_added'])
```

```python
# convert release_year column to string
dataset['release_year'] = dataset['release_year'].astype('int64')
```

```python
# print out the data type of each column
dataset.dtypes
```

```python
dataset
```

```python
# How to capitalise the first word of each element in a list
dataset.columns = [a.title() for a in dataset.columns]
```

### Conclusion
From careful observation of the dataset;
