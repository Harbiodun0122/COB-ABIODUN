### Phase 1 Project 1
Create a csv dataset using python , pandas and any public api

For this project, want to to create a csv dataset that contains a currency, type of currency and the exchange rate in dollar.
The public api used is api.coincap.io

```python
# import the required libraries
import requests
import pandas as pd
```

```python
# web url
url = 'https://api.coincap.io/v2/rates'
```

```python
# specify the headers
headers = {
    'Accept':'application/json',
    'Content-Type': 'application/json'
}
```

```python
# send a request to the url
response = requests.request('GET', url, headers=headers, data={})
```

```python
# convert the response to json
data = response.json()
```

```python
# convert the json to a pandas dataframe
dataset = pd.DataFrame(data['data'])
```

```python
# print out the dataset
dataset
```

### Data Cleaning

```python
# drop the currencySymboel column since it's not needed
dataset = dataset.drop(columns='currencySymbol', axis=0)
```

```python
dataset
```

```python
# print out the datatype of each column in the dataset
dataset.dtypes
```

```python
# convert the rateUsd column to a numerical column
dataset['rateUsd'] = pd.to_numeric(dataset['rateUsd'])
```

```python
# print out each column datatype
dataset.dtypes
```

```python
# round up the rateUsd column to 4 decimal places
dataset['rateUsd'] = dataset['rateUsd'].apply(lambda x: round(x, 4))
```

```python
dataset
```

```python
# save the dataframe to a csv file
dataset.to_csv('Currency Exchange Rate.csv')
```
